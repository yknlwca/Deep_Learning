```python
okt = Okt()
# 한국어 불용어 리스트
stopwords = ['을', '를', '이', '가', '은', '는']
```

```python
okt = Okt()
# 한국어 불용어 리스트
stopwords = ['을', '를', '이', '가', '은', '는']
# 한국어 텍스트의 불용어 제거 함수 생성
def remove_stopwords(word_list):
    tokens = okt.morphs(' '.join(word_list))  # 형태소 분석
    tokens = [token for token in tokens if token not in stopwords]  # 불용어 제거
    return tokens

combined_lan['sentence'] = combined_lan['sentence'].apply(remove_stopwords)
combined_lan
```

```python
# 역토큰화 과정
def detokenize(tokens):
    return ' '.join(tokens)

combined_lan['sentence'] = combined_lan['sentence'].apply(detokenize)
combined_lan
```

```python
# 양이 적어 max_features = None
vectorizer = TfidfVectorizer(max_features=None)
X = vectorizer.fit_transform(combined_lan['sentence'])
print('TF-IDF 행렬의 크기 :',X.shape)
```

```python
lda_model = LatentDirichletAllocation(n_components=10, learning_method='online', random_state=1, max_iter=1)
```

```python
lda_top = lda_model.fit_transform(X)
terms = vectorizer.get_feature_names_out()
```

```python
def get_topics(comp, fn, n=5):
    for idx, topic in enumerate(comp):
        print("Topic %d:" % (idx+1), [(fn[i], topic[i].round(2)) for i in topic.argsort()[:-n - 1:-1]])

get_topics(lda_model.components_, terms)
```
